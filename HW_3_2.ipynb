**1. Завантаження та огляд**

```python
import pandas as pd
ds = pd.read_csv(r'C:\Users\Олександра\Downloads\tips.csv')

print('Перші 5 рядків')
display(ds.head())

print(f'Розмір датасету: {ds.shape}\n')

print(f'Назви колонок: {list(ds.columns)}\n')

print('Інформація про тип даних')
display(ds.info())
```

**2. Базова вибірка даних**

```python
total_bill = ds[['total_bill']]
print(f'Колонка total_bill')
display(total_bill)

coop = ds[['total_bill', 'tip']]
print(f'\n Колонки total_bill та tip разом')
display(coop)

print('\n Перші 10 рядків')
display(ds.head(10))

print('\n Рядок з індексом 5')
display(ds.loc[5])
```

**3. Фільтрація даних**

```python
high_bill = ds[ds['total_bill']>20]
print('Рахунки більше 20 доларів')
print(f'Кількість записів: {high_bill.shape}')
display(high_bill.head())

female = ds[ds['sex']=='Female']
print('\nВідвідувачі - жінки')
print(f'Кількість записів: {female.shape}')
display(female.head())

dinner_bill = ds[ds['time']=='Dinner']
print('\nРахунки за вечерю')
print(f'Кількість записів: {dinner_bill.shape}')
display(dinner_bill.head())

dinner_high_bill = ds[(ds['time']=='Dinner') & (ds['total_bill']>30)]
print('\nРахунки за вечерю дорожче 30 доларів')
print(f'Кількість записів: {dinner_high_bill.shape}')
display(dinner_high_bill.head())
```

**4. Описова статистика**

```python
average = ds['total_bill'].mean()
tip_mx = ds['tip'].max()
empty = ds.isnull().sum().sum()
bill_time = ds['time'].value_counts()
print(f'Середній розмір рахунку: {average.round(2)}')
print(f'Максимальні чайові: {tip_mx}')
print(f'Кількість пропущених значень у всіх колонках: {empty}')
print(f'Кількість рахунків за обіди і за вечері:\n{bill_time}')
```

**5. Статистика із групуванням**

```python
avr_sex = ds.groupby('sex')['total_bill'].mean()
avr_tips = ds.groupby('day')['tip'].mean()
number_day = ds.groupby('day')['size'].value_counts()
print(f'Середній рахунок за статтю:{avr_sex.round(2)}')
print(f'\nСередні чайові за днем тижня:{avr_tips.round(2)}')
print('\nКількість відвідувачів у кожен день')
display(number_day)
```

**6. Створення нової колонки**

```python
ds['tip_percentage'] = (ds['tip']/ds['total_bill'])*100
display(ds.head(10))
```

**7. Сортування**

```python
bill_sorted = ds.sort_values('total_bill')
tip_sorted = ds.sort_values('tip', ascending = False)
top_five = ds.nlargest(5,'tip')
print('Сортування за зростанням розміру рахунку')
display(bill_sorted)
print('\nСортування за спаданням розміру чайових')
display(tip_sorted)
print('\nТоп 5 найбільших чайових')
display(top_five)
```
