```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

***#1***

```python
x = sp.symbols('x')
fun = 10 * x - 5
derivative_fun = sp.diff(fun, x)
derivative_fun
```

```python
x = np.linspace(0,10)
func = 10*x - 5
dfunc = x*0 + 10

plt.figure(figsize=(10, 5))
plt.plot(x, func, label='f(x) = 10x - 5', color='red')
plt.plot(x, dfunc, label="f'(x) = 10", color='orange')
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

***#2***

```python
a = sp.symbols('a')
func_a = sp.sin(a)
derivative_func_a = sp.diff(func_a, a)
derivative_func_a
```

```python
a = sp.symbols('a')
func_a = sp.sin(a)
dfunc_a = sp.cos(a)

fa_num = sp.lambdify(a, func_a, 'numpy')
dfa_num = sp.lambdify(a, dfunc_a, 'numpy')

x_vals = np.linspace(-2*np.pi, 2*np.pi)

y_vals = fa_num(x_vals)
dy_vals = dfa_num(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(a) = sin(a)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(a) = cos(a)", color='violet')
plt.title("Графік функції та її похідної для f(a) = sin(a)")
plt.xlabel("a")
plt.ylabel("f(a)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
b = sp.symbols('a')
func_b = sp.ln(b)
derivative_func_b = sp.diff(func_b, b)
derivative_func_b
```

```python
b = sp.symbols('b')
func_b = sp.ln(b)
dfunc_b = 1/b

fb_num = sp.lambdify(b, func_b, 'numpy')
dfb_num = sp.lambdify(b, dfunc_b, 'numpy')

b_vals = np.linspace(1,10)

yb_vals = fb_num(b_vals)
dyb_vals = dfb_num(b_vals)

plt.figure(figsize=(10, 5))
plt.plot(b_vals, yb_vals, label='f(b) = ln(a)', color='green')
plt.plot(b_vals, dyb_vals, label="f'(b) = 1/b", color='navy')
plt.title("Графік функції та її похідної для f(b) = ln(b)")
plt.xlabel("b")
plt.ylabel("f(b)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = sp.symbols('x')
fun = x**2 - 5*x + 1
derivative_fun = sp.diff(fun, x)
derivative_fun
```

```python
x = np.linspace(0,100)
fun = x**2 - 5*x + 1
dfun = 2*x - 5

plt.figure(figsize=(10, 5))
plt.plot(x, fun, label='f(x) = x^2 − 5x + 1', color='navy')
plt.plot(x, dfun, label="f'(x) = 2x - 5", color='orange')
plt.title("Графік функції та її похідної обсягу виробництва")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
