**1. Функція з параметром**

```python
def show_menu(dishes):
  i = 1
  print(f"=== МЕНЮ ===")
  for dish in dishes:
    print(f"{i}. {dish}")
    i = i+1
  print("=============")

new_menu = ["Піца", "Паста", "Салат"]
show_menu(new_menu)
```

**3. Функція з поверненням значення**

**4. Функція пошуку максимального значення**

```python
def find_max(numbers):
    return max(numbers)

list = [15, 42, 8, 23, 56, 19]
value_max = find_max(list)
print(f"Найбільше число - {value_max}")
```
