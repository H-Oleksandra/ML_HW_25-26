**1. Списки - базові операції**

```python
temperatures = [22,25,19,23,26,20,21,22]
```

```python
temperatures[0]
```

```python
temperatures[-1]
```

```python
# додавання температури 24 в кінець
temperatures.append(24)
```

```python
temperatures
```

```python
# додавання температури 24 на початок
temperatures.insert(0,24)
temperatures
```

```python
# видалення зайвих 24 - виправлення помилки виконання попереднього коду
temperatures.remove(24)
temperatures
```

```python
temperatures.remove(19)
temperatures
```

```python
len(temperatures)
```

```python
sum(temperatures)
```

```python
max(temperatures)
```
