**1. Створення масивів**

```python
import numpy as np
arr = np.array([5,15,25,35,45])
arr_zero = np.zeros(7)
arr_one = np.ones(6)
arr_numb = np.arange(10,20)
print(f"Масив зі списку: {arr}")
print(f"Масив нулів: {arr_zero}")
print(f"Масив одиниць: {arr_one}")
print(f"Масив з числами від 10 до 19: {arr_numb}")
```

**2. Базові операції**

```python
arr = np.array([12,24,36,48,60])
print(f"Множення на 2: {arr * 2}")
print(f"Додавання 10: {arr + 10}")
print(f"Піднесення до квадрату: {arr ** 2}")
print(f"Квадратний корінь: {np.sqrt(arr)}")
```

**4. Індексанція**

```python
arr = np.array([15,22,35,44,58,67])
print(f"Перший елемент: {arr[0]}")
print(f"Останній елемент: {arr[-1]}")
print(f"Елементи з індексами 1-3: {arr[1:4]}")
print(f"Елементи більші за 40: {arr[arr > 40]}")
```

**3. Статистичні функції**

```python
data = np.array([18,27,33,41,52,68,79])
print(f"Середнє значення: {np.mean(data)}")
print(f"Мінімум: {np.min(data)}")
print(f"Максимум: {np.max(data)}")
print(f"Сума елементів: {np.sum(data)}")
```

**5. Двовимірні масиви**

```python
matrix = np.array([[2,4,6],
                  [8,10,12],
                  [14,16,18]])
print(f"Форма матриці: {matrix.shape}")
print(f"Перший рядок: {matrix[0,:]}")
print(f"Друга колонка: {matrix[:,1]}")
print(f"Сума всіх елементів: {np.sum(matrix)}")
print(f"Середнє всіх елементів: {np.mean(matrix)}")
```
