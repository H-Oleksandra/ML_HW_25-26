```python
import pandas as pd
ds = pd.read_csv(r'C:\Users\Олександра\Downloads\tips.csv')
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
```

**1. Гістограма розподілу рахунків**

```python
ds['total_bill'].hist(bins=20)
plt.title('Розподіл рахунків')
plt.xlabel('Сума рахунку')
plt.ylabel('Кількість рахунків')
plt.show()
```

**2. Box Plot чайових за дними тижня**

```python
ds.boxplot(column='tip', by='day')
plt.title('Розподіл чайових')
plt.suptitle('')
plt.xlabel('День')
plt.ylabel('Сума чайових')
plt.show()
```

Найвище медіанне (середнє) значення чайових у неділю.

**3. Scatter Plot: рахунок vs чайові**

```python
ds.plot.scatter(x='total_bill', y='tip', alpha=0.5)
plt.title('Рахунки vs чайові')
plt.xlabel('Рахунки')
plt.ylabel('Чайові')
plt.show()
```

**4. Bar Chart: середній рахунок за часом**

```python
ds.groupby('time')['total_bill'].mean().plot(kind='bar')
plt.title('Cередній рахунок за часом')
plt.xlabel('Час')
plt.xticks(rotation=0)
plt.ylabel('Середній рахунок')
plt.show()
```

**5. Інтерактивна діаграма розсіювання**

```python
diag=px.scatter(ds, x='total_bill', y='tip', color='sex',
                title='Рахунок та чайові за статтю',
                labels={'total_bill': 'Рахунок', 'tip': 'Чайові','sex': 'Стать'})
diag.show()
```

**6. Інтерактивний plot box: чайові за днями та статтю**

```python
diag_box=px.box(ds, x='day', y='tip', color='sex',
                title='Розподіл чайових за днями та статтю',
                labels={'day': 'День', 'tip': 'Чайові','sex': 'Стать'})
diag_box.show()
```
