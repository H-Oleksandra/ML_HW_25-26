```python
import numpy as np
```

***Завдання 1: Створення матриць та векторів***

```python
vector = np.random.rand(10)
matrix = np.array([[11,14,17],
                  [20,23,26],
                  [29,32,35]])
print(f'Вектор та його розмір:\n{vector}\n{vector.shape}')
print(f'Матриця та її розмір:\n{matrix}\n{matrix.shape}')
```

***Завдання 2: Операції над векторами та матрицями***

```python
vector_add = vector + 10
matrix_mult = matrix * 2
vector_sqrt = vector ** 2

print(f'Додавання вектора\n {np.array2string(vector_add, precision=2, suppress_small=True)}')
print(f'Множення матриці\n {matrix_mult}')
print(f'Вектор квардратів\n {vector_sqrt}')
```

***Завдання 3: Множення матриць***

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

result_matrix = np.dot(B, A)
result_matrix
```

```python
C = np.array([[0], [1]])
result_matrix_C = np.dot(result_matrix, C)
result_matrix_C
```

```python
D = np.array([[0,1], [1,0]])
result_matrix_D = np.dot(result_matrix, D)
result_matrix_D
```
