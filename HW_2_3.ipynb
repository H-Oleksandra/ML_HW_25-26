**1. Основи роботи з умовними конструкціями (if-elif-else)**

```python
cooking_time = 5
if cooking_time<4:
    print("Яйце рідке")
elif cooking_time<7:
    print("Яйце всмятку")
elif cooking_time<10:
    print("Яйце вкруту")
else:
    print("Яйце переварене")
```

**2. Перевірка оцінки**

```python
grade = 72
if grade>=90:
    print("Відмінно")
elif grade>=75:
    print("Добре")
elif grade>=60:
    print("Задовільно")
else:
    print("Незадовільно")
```

**3. Перевірка на наявність**

```python
food_list = ["молоко","яйця","масло","сир"]
presence = "Яйця є в холодильнику" if "яйця" in food_list else "Яєць немає"
print(presence)
if "хліб" in food_list:
  print("Хліб є")    
```
