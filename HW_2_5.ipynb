**1. Цикл while**

```python
counter = -3
while counter<5:
    counter += 1
    print(counter)
```

**2. List comprehension - прості операції**

```python
numbers = [11,4,9,16,7,5]
squares = [a**2 for a in numbers]
even = [b for b in numbers if b % 2 == 0]
multiply = [c*10 for c in numbers]
print(f"Квадрати чисел: {squares}")
print(f"Парні числа: {even}")
print(f"Помножені на 10 числа: {multiply}")
```

**3. List comprehension - комбінована логіка**

```python
numbers = list(range(1,11))
number = [i*2 if i%2==0 else i-1 for i in numbers]
print(f"Новий список: {number}")
```
